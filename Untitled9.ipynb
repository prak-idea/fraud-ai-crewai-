```python
from crewai.llm import BaseLLM
from groq import Groq

client = Groq()

class GroqLLM(BaseLLM):
    def __init__(self, model="llama-3.3-70b-versatile"):
        super().__init__(model=model)
        self.model = model

    def call(self, prompt, **kwargs):
        if not isinstance(prompt, str):
            prompt = str(prompt)

        response = client.chat.completions.create(
            model=self.model,
            messages=[{"role": "user", "content": prompt}],
            temperature=0.2
        )
        return response.choices[0].message.content
```

```python
llm = GroqLLM()
```

```python
from crewai import Agent

researcher = Agent(
    role="Fraud Researcher",
    goal="Find deep insights about financial frauds",
    backstory="Expert in global financial crime investigation",
    llm=llm,
    verbose=True
)

analyzer = Agent(
    role="Pattern Analyzer",
    goal="Identify fraud patterns",
    backstory="Data scientist",
    llm=llm,
    verbose=True
)

reporter = Agent(
    role="Report Generator",
    goal="Generate structured report",
    backstory="Financial analyst",
    llm=llm,
    verbose=True
)
```

```python
from crewai import Task

task1 = Task(
    description="Research latest crypto scams in 2025",
    expected_output="Detailed list of latest crypto scams, techniques, and real-world examples",
    agent=researcher
)

task2 = Task(
    description="Analyze fraud patterns from the researched scams",
    expected_output="Clear analysis of common fraud patterns, attacker strategies, and behavioral trends",
    agent=analyzer
)

task3 = Task(
    description="Generate a detailed structured fraud intelligence report",
    expected_output="Well-structured report including insights, patterns, risks, and conclusions",
    agent=reporter
)
```

```python
from crewai import Crew

crew = Crew(
    agents=[researcher, analyzer, reporter],
    tasks=[task1, task2, task3],
    verbose=True
)
```

```python
from crewai.tools import BaseTool
from duckduckgo_search import DDGS

class SearchTool(BaseTool):
    name: str = "Web Search Tool"
    description: str = "Search the web for latest fraud cases and information"

    def _run(self, query: str) -> str:
        results = []
        with DDGS() as ddgs:
            for r in ddgs.text(query, max_results=5):
                results.append(r["title"] + " - " + r["body"])
        return "\n".join(results)
```

```python
search_tool = SearchTool()
```
